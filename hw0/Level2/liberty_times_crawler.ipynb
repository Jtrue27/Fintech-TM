#### Observe that we can get news of a day with the link format:
http://news.ltn.com.tw/list/newspaper/politics/20181231
#### We need a list of date string in the date range we want.

```python
from datetime import datetime, timedelta

start_date = "2018-07-01"
stop_date = "2018-12-31"

start = datetime.strptime(start_date, "%Y-%m-%d") # datetime.strptime(date_string, format) return datetime
stop = datetime.strptime(stop_date, "%Y-%m-%d")

dates = list() #dates=[]
while start <= stop:
    dates.append(start.strftime('%Y%m%d')) # Return a string representing the date and time
    start = start + timedelta(days=1) 
```

#### Now write a function to parse the HTML response, return the data we want(title, content, ...etc).

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
def process_document(document, date):
    
    nodes = document.select('ul.list > li') # 選擇ul的class list 底下的li
    data = list()

    for li in nodes:

        # check if is empty element
        if li.select_one('a') == None: # 沒有<a></a> 是我們不要的
            continue # select_one = finds only the first tag that matches a selector

        # get link
        li_link = 'http://news.ltn.com.tw/' + li.select_one('a')['href'] #把網址取出來

        # request for document
        li_res = requests.get(li_link) 
        li_doc = bs(li_res.text, 'lxml')

        # get date
        li_date = datetime.strptime(date, "%Y%m%d").strftime('%Y-%m-%d') # 轉回去字串 "2018-07-01"

        #get title
        li_title = li.select_one('p').get_text() #取得 title

        #get content
        li_content = ""
        for ele in li_doc.select('div.text > p'): # 在文章中的div.text下的p
            if not 'appE1121' in ele.get('class', []): # 不要想看更多新聞
                
                li_content += ele.get_text() 

        # append new row
        data.append({
            'date' : li_date,
            'title': li_title,
            'link' : li_link,
            'content' : li_content,
            'tags' : []
        })
    return data
```

#### Crawl over the news on the site, store the data in variable "all_data" .

```python
cnt = 0
all_data = list()
for date in dates:
    print('start crawling :', date)
    res = requests.get('https://news.ltn.com.tw/list/newspaper/politics/' + date) # ex: http://news.ltn.com.tw/list/newspaper/politics/20181231
    doc = bs(res.text, 'lxml') # lxml為處理HTML的編譯器套件，該網頁經過BeautifulSoup內lxml編輯器解析的結果
    data = process_document(doc, date)
    all_data += data
```
