```python
from bs4 import BeautifulSoup
import requests
import re

#Get News Title: title_list, Url: url_list, time: time_list
#Input: pg: page to crawl, url: the first page url
def news_crawler(pg, url):
    title_list = []
    url_list = []
    time_list = []
    for page in range(1, pg+1):
        #取出搜尋頁面之店家基本資料
        news_url = url + 'page/' + str(page) + '/'
        res = requests.get(news_url)
        soup = BeautifulSoup(res.text, 'html.parser')
        sel = soup.find_all('article')
        for article in sel:
            context = article.find('h3')
            url_list.append(context.findAll('a', href=True, text=True)[0]['href'])
            for text in context.find('a'):
                title_list.append(text)
                print('========' + text + '========')
            update_time = article.find('span', class_ = 'mh-meta-date updated')
            time_list.append(str(update_time)[64:74])
        print('=======page:' + str(page) + '=======')
    return title_list, url_list, time_list
        

        
```

```python
#Get News Title: title_list, Url: url_list, time: time_list
#Input: pg: page to crawl, url: the first page url
def content_crawler(url_list):
    content_list = []
    count = 0
    for url in url_list:
        count += 1
        res = requests.get(url)
        soup = BeautifulSoup(res.text, 'html.parser')
        sel = soup.find_all('article')
        for article in sel:
            content_text = []
            added_content = ''
            context = article.find('div', class_ = 'entry-content mh-clearfix')
        try:
            for content in context:
                content_text.append(content.text)
            for i in range(0,len(content_text)-3):
                added_content += content_text[i]
            content_list.append(added_content.replace(u'\xa0', u''))
            print('Finish crawling ' + str(count) + ' news content')
        except:
            content_list.append('WRONG CONTENT')
            print('Check again on: ' + str(url))
    return content_list
```
